```python
import matplotlib.pyplot as plt
import numpy as np

# Dados
grupos_idade = [
    '100+', '95-99', '90-94', '85-89', '80-84', '75-79', '70-74', '65-69', '60-64',
    '55-59', '50-54', '45-49', '40-44', '35-39', '30-34', '25-29', '20-24', '15-19',
    '10-14', '5-9', '0-4'
]

pop_feminina = [
    37, 227, 741, 1624, 2791, 4062, 6079, 7936, 9923, 10783, 11084, 11653, 14070,
    13923, 13633, 13338, 12729, 10776, 10261, 10226, 9405
]

pop_masculina = [
    12, 88, 335, 899, 1674, 2856, 4459, 6326, 8343, 9187, 9778, 10747, 13233,
    13337, 13144, 13123, 12771, 10937, 10569, 10839, 9878
]

# Inverta a ordem dos dados femininos para que a pirâmide seja correta
grupos_idade = grupos_idade[::-1]

# Inverta a ordem dos dados masculinos para que a pirâmide seja correta
pop_masculina = pop_masculina[::-1]

# Inverta a ordem dos dados femininos para que a pirâmide seja correta
pop_feminina = pop_feminina[::-1]

# Cores mais suaves
cor_feminina = '#FFB6C1'  # Rosa claro
cor_masculina = '#ADD8E6'  # Azul claro

# Crie o gráfico de pirâmide
fig, ax = plt.subplots()

# Barra para a população feminina
ax.barh(grupos_idade, pop_feminina, color=cor_feminina, label='Feminino')

# Barra para a população masculina (com valores negativos para inverter a direção)
ax.barh(grupos_idade, [-x for x in pop_masculina], color=cor_masculina, label='Masculino')

# Remova os números do eixo X
ax.set_xticks([])
# Ajuste de estilo
#ax.set_xlabel('População')
#ax.set_ylabel('Grupo de Idade')
ax.set_title('Distribuição da População por Idade e Gênero')
#ax.legend()

# Exiba o gráfico
plt.show()
```

```python
import plotly.graph_objects as go

# Dados
grupos_idade = [
    '100+', '95-99', '90-94', '85-89', '80-84', '75-79', '70-74', '65-69', '60-64',
    '55-59', '50-54', '45-49', '40-44', '35-39', '30-34', '25-29', '20-24', '15-19',
    '10-14', '5-9', '0-4'
]

pop_feminina = [
    37, 227, 741, 1624, 2791, 4062, 6079, 7936, 9923, 10783, 11084, 11653, 14070,
    13923, 13633, 13338, 12729, 10776, 10261, 10226, 9405
]

pop_masculina = [
    12, 88, 335, 899, 1674, 2856, 4459, 6326, 8343, 9187, 9778, 10747, 13233,
    13337, 13144, 13123, 12771, 10937, 10569, 10839, 9878
]

# Inverta a ordem dos dados femininos para que a pirâmide seja correta
grupos_idade = grupos_idade[::-1]

# Inverta a ordem dos dados masculinos para que a pirâmide seja correta
pop_masculina = pop_masculina[::-1]

# Inverta a ordem dos dados femininos para que a pirâmide seja correta
pop_feminina = pop_feminina[::-1]

# Cores mais suaves
cor_feminina = '#FFB6C1'  # Rosa claro
cor_masculina = '#ADD8E6'  # Azul claro

# Crie o gráfico de pirâmide
fig = go.Figure()

# Adicione barras para a população feminina
fig.add_trace(go.Bar(
    y=grupos_idade,
    x=pop_feminina,
    orientation='h',
    marker=dict(color=cor_feminina),
    name='Feminino'
))

# Adicione barras para a população masculina
fig.add_trace(go.Bar(
    y=grupos_idade,
    x=[-x for x in pop_masculina],  # valores negativos para inverter a direção
    orientation='h',
    marker=dict(color=cor_masculina),
    name='Masculino'
))

# Ajuste de layout
fig.update_layout(
    barmode='overlay',
    xaxis=dict(visible=False),
    yaxis=dict(title='Grupo de Idade'),
    title='Distribuição da População por Idade e Gênero',
    legend=dict(x=0, y=1),
)

# Adicione tooltips interativos
fig.update_traces(
    hovertemplate='%{y}<br>Homens: %{x}<br>Mulheres: %{x}'
)

# Exiba o gráfico
fig.show()
```
